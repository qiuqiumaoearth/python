假设在空池塘中放入一颗水藻，该类水藻会每周长出三颗新的水藻，问十周后，池塘中有多少颗水藻？

```python
x = 1 # 初始条件：第一周水藻数量
times = 1 # 迭代次数
while times < 10: # 迭代过程
    y = 3 * x
    x = x + y
    times += 1
    print("第%d周的水藻数量:%d" % (times, x))
```

求解方程x^3+e^x/2+5x-6=0的根。 建立迭代格式： x=(6-x^3-(e^x)/2)/5

```python
import math
x = 0 # 初始条件
for i in range(100): # 控制总的迭代次数
    x = (6 - x**3 - (math.e**x)/2.0)/5.0 
    print(str(i)+":"+str(x))
```

```python
x = 0 # 初始条件
delta = 0.00001 # 控制退出条件
times = 0 # 用来显示迭代次数
while True: # 条件为True，如果没有别的退出手段，while循环将会无限进行下去
    x_old = x
    x = (6 - x**3 - (math.e**x)/2.0)/5.0 
    print(str(times)+":"+str(x))
    times += 1
    if abs(x - x_old) < delta:
        break # 如果符合退出条件，则直接退出循环
```
